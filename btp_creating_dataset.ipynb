```python
def clear_sent(text):
    res=[]
    text = text.replace("gonna", "going to")
    text = text.lower()
    doc = nlp(text)
    for token in doc:
        x = str(token.text)
        x = x.replace(" ", "")
        if(x=="'s" or x=="s"): res.append("is")
        elif(x=="'re" or x=="re"): res.append("are")
        elif(x=="'m" or x=="m"): res.append("am")
        elif(x=="n't"): res.append("not")
        elif(x=="ca"): res.append("can")
        elif(x=="wo"): res.append("will")
        elif(x=="i"): res.append("me")
        elif(x=="'ve" or x=="ve"):res.append("have")
        elif(x=="'ll" or x=="ll"):res.append("will")
        elif(x=="'d" or x=="d"): res.append("would")
        elif (x=="yea" or x=="yeah"): res.append("yes")
        elif(x=="" or x=="yyy" or x=="xxx" or x=="um" or x=="oh" or x=="uh" or x=="hm" or x=="hmm" or x=="huh" or x=="\'" or x=='\"' or x=="’"): pass
            
            
        else: res.append(token.lemma_)
    return res
```

```python
bliss_id = pd.read_csv("bliss_id_word.csv")[:-1]
bliss_id
blis_word_list = {str(bliss_id["gloss"][i]):int(bliss_id["BCI reference number"][i]) for i in range(len(bliss_id))}
blis_word_list
```

```python
def expand_list(wrd_list):
    l ={}
    for i in wrd_list:
        for j in i.split(","):
            if j not in l: l[j] = wrd_list[i];
    return l
l = expand_list(blis_word_list)
l
```

```python
def clean_bliss_word(wrd):
    x = wrd.split("_(")[0]
    x = x.split("-(")[0]
    try: x = x.split("(")[0]+x.split("(")[1][:-1]
    except: pass
    return x;
```

```python
bliss_word_cleaned = {clean_bliss_word(i):l[i] for i in l}
```

```python
bliss_word_dict = expand_list(bliss_word_cleaned)
```

```python
from tqdm import tqdm
```

```python
punct_marks = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
eng = []; bliss = []
for i in tqdm(range(len(child))):
    c = 0;
    bb = ""
    for j in clear_sent(child["text"][i]):
        if j in punct_marks: 
            bb += j+" ";
            continue
        if j not in bliss_word_dict:
            c+=1;
            bb += "-999999 "
        else: bb += str(str(bliss_word_dict[j])+" ")
    
    eng.append(child["text"][i])
    bliss.append(bb)
```

```python
punct_marks = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
eng = []; bliss = []
for i in tqdm(range(len(data))):
    c = 0;
    bb = ""
    for j in clear_sent(data["dialog_text"][i]):
        if j in punct_marks: bb += j+" ";
        if j not in bliss_word_dict:
            c+=1;
            bb += "-9999 "
        else: bb += str(str(bliss_word_dict[j])+" ")
    
    eng.append(data["dialog_text"][i])
    bliss.append(bb)
```
